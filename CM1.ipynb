# **Подключаем библиотеки**

```python
import random
import math
import seaborn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tabulate
```

# **Ввод данных**

```python
a = 4
b = 7
l = 2
m = 0
s = 4
```

```python
BINS=16          
DEG=6           
N=10**DEG


def R() -> float:
    R = 0
    while R == 0:
        R = random.random()
    return R

```

# **Равномерное распределение**



```python
def equitable_distribution(a: int, b: int, n: int) -> list:
    dist = [0] * n
    for i in range(n):
        dist[i] = a + R()*(b-a)
    return dist
```

```python
d1 = equitable_distribution(a, b, N)
seaborn.displot(d1, bins=BINS, color = "Violet")
plt.xlabel('Равномерное распределение')
#seaborn.displot(d1, kind='kde')
```

```python
nn=1000
print('Равномерное распредение с N = 1000')
equitable_distribution(a,b,nn)
```

# **Экспоненциальное распределение**

```python
def exponential_distribution(l: int, n: int) -> list:
    dist = [0] * n
    for i in range(n):
        dist[i] = (-1) * math.log(R())/ l
    return dist
```

```python
d2 = exponential_distribution(l, N)
seaborn.displot(d2, bins=BINS, color = "yellow")
#seaborn.displot(d2, kind='kde')
```

```python
print('Экспоненциальное распредение с N = 1000')
exponential_distribution(l, nn)
```

# **Нормальное распределение**

```python
def normal_distribution(m: int, s:int, n: int) -> list:
    dist = [0] * n
    for i in range(n):
        R12 = 0
        for j in range(12):
            R12 += R()
        dist[i] = m + s * (R12 - 6)
    return dist 
```

```python
d3 = normal_distribution(m, s, N)
seaborn.displot(d3, bins=BINS, color = "blue")
#seaborn.displot(d3, kind='kde')
```

```python
print('Нормальное распредение с N = 1000')
normal_distribution(m, s, nn)
```

# **Функции для математического ожидания и дисперсии**

```python
def M(dist: list) -> float:
    return np.mean(dist)

def D(dist: list) -> float:
    return np.var(dist, ddof=1)
```

**Размеры выборок**

```python
NS = [10,20,50,100,1000]
```

```python
dists = [d1, d2, d3]
est = [[None for i in range(len(NS))] for j in range(len(dists) * 2)]

for i in range(len(dists)):
    for j in range(len(NS)):
        est[2*i][j] = M(random.choices(dists[i], k=NS[j])) 
        est[2*i+1][j] = D(random.choices(dists[i], k=NS[j]))
        # random.choices(list, k) - выберет из list k элементов случайным образом (могут повторяться)

true_est = [(a+b)/2, ((b-a)**2)/12, 1/l, 1/(l*l), m, s]
```

# **Вывод таблицы**

```python
table = []
dist_names = ['Равномерное', 'Экспоненциальное', 'Нормальное']
h = ["N(i)"] + NS + [True]

for i in range(len(dists)):
    table.append([f'М({dist_names[i]}_распределение)'] + est[2*i] + [true_est[2*i]])
    table.append([f'Д({dist_names[i]}_распределение)'] + est[2*i+1] + [true_est[2*i+1]])

print(tabulate.tabulate(table, headers=h, tablefmt="fancy_grid", floatfmt=".3f"))
```

```python
nplist = []
hh = [f'N{i+1}' for i in range(len(h)-1)]
hh[-1] = 'N' # подписи для оси Х

# загоняем все данные в датафрейм в ненормализованном виде
for i in range(len(table)):
    for j in range(1, len(table[i])):
        nplist.append([table[i][0], hh[j-1], table[i][j]])

data = pd.DataFrame(np.array(nplist), columns=['est-dist', 'sample', 'value'])
data = data.astype({'value': np.double }) # нужно дать нормальный тип столбцу, который будет вертикальной осью
```

# **Каждая оценка на отдельном графике**

```python
g = seaborn.FacetGrid(data, col='est-dist', col_wrap=2, height=5.5)
g.map(seaborn.lineplot, 'sample', 'value')
```
